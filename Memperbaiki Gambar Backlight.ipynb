```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#202331161_Agustne L B. Nikijuluw
img = cv2.imread('C:/Users/62852/OneDrive/Gambar/Backlight.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
#202331161_Agustne L B. Nikijuluw
brightness_factor = 50
bright_img = cv2.add(gray, brightness_factor)
alpha = 1.5
beta = 0
contrast_img = cv2.convertScaleAbs(gray, alpha=alpha, beta=beta)
combined_img = cv2.convertScaleAbs(bright_img, alpha=alpha, beta=0)
```

```python
#202331161_Agustne L B. Nikijuluw
plt.figure(figsize=(15, 12))
plt.subplot(2, 3, 1), plt.imshow(img_rgb), plt.title('Original Image'), plt.axis('off')
plt.subplot(2, 3, 2), plt.imshow(gray, cmap='gray'), plt.title('Gray'), plt.axis('off')
plt.subplot(2, 3, 3), plt.imshow(bright_img, cmap='gray'), plt.title('Gray yang dipercerah '), plt.axis('off')
plt.subplot(2, 3, 4), plt.imshow(contrast_img, cmap='gray'), plt.title('Gray yang diperkontras'), plt.axis('off')
plt.subplot(2, 3, 5), plt.imshow(combined_img, cmap='gray'), plt.title('Dipercerah + Diperkontras'), plt.axis('off')
plt.tight_layout()
plt.savefig('backlight_correction_results.png')
plt.show()
```
