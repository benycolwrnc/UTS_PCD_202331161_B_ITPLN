```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#202331161_Agustne L B. Nikijuluw
img = cv2.imread('C:/Users/62852/OneDrive/Gambar/202331161_Agustne.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
#202331161_Agustne L B. Nikijuluw
lower_red = np.array([0, 70, 50])
upper_red = np.array([10, 255, 255])
lower_green = np.array([40, 40, 40])
upper_green = np.array([80, 255, 255])
lower_blue = np.array([100, 50, 50])
upper_blue = np.array([130, 255, 255])
```

```python
#202331161_Agustne L B. Nikijuluw
mask_red = cv2.inRange(img_hsv, lower_red, upper_red)
mask_green = cv2.inRange(img_hsv, lower_green, upper_green)
mask_blue = cv2.inRange(img_hsv, lower_blue, upper_blue)
```

```python
#202331161_Agustne L B. Nikijuluw
red_img = cv2.bitwise_and(img, img, mask=mask_red)
green_img = cv2.bitwise_and(img, img, mask=mask_green)
blue_img = cv2.bitwise_and(img, img, mask=mask_blue)
```

```python
#202331161_Agustne L B. Nikijuluw
red_img = cv2.cvtColor(red_img, cv2.COLOR_BGR2RGB)
green_img = cv2.cvtColor(green_img, cv2.COLOR_BGR2RGB)
blue_img = cv2.cvtColor(blue_img, cv2.COLOR_BGR2RGB)
```

```python
#202331161_Agustne L B. Nikijuluw
contrast_img = cv2.convertScaleAbs(img_rgb, alpha=1.5, beta=30)
```

```python
#202331161_Agustne L B. Nikijuluw
cv2.imwrite('citra_kontras.jpg', cv2.cvtColor(contrast_img, cv2.COLOR_RGB2BGR))
cv2.imwrite('citra_merah.jpg', cv2.cvtColor(red_img, cv2.COLOR_RGB2BGR))
cv2.imwrite('citra_hijau.jpg', cv2.cvtColor(green_img, cv2.COLOR_RGB2BGR))
cv2.imwrite('citra_biru.jpg', cv2.cvtColor(blue_img, cv2.COLOR_RGB2BGR))
```

```python
#202331161_Agustne L B. Nikijuluw
plt.figure(figsize=(12, 8))
    
plt.subplot(2, 2, 1), plt.title('CITRA KONTRAS'), plt.imshow(contrast_img), plt.axis('off')
plt.subplot(2, 2, 2), plt.title('BIRU'), plt.imshow(blue_img), plt.axis('off')
plt.subplot(2, 2, 3), plt.title('MERAH'), plt.imshow(red_img), plt.axis('off')
plt.subplot(2, 2, 4), plt.title('HIJAU'), plt.imshow(green_img), plt.axis('off')
```

```python
#202331161_Agustne L B. Nikijuluw
plt.figure(figsize=(10,5))
color = ('b', 'g', 'r')
for i, col in enumerate(color):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color=col)
plt.title('Histogram RGB')
plt.xlim([0, 256])
plt.tight_layout()
plt.show()
```
